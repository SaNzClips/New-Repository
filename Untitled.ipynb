```python
import pandas as pd
df_incidents = pd.read_csv('https://github.com/shihao-wen/IBM-Data-Science-Professional-Certificate/blob/master/6.%20Data%20Visualization/Final%20Assignment/Police_Department_Incidents_-_Previous_Year__2016_.csv?raw=true')
print('Dataset downloaded and read into a pandas dataframe!')
df_incidents.head()
```

```python
# get the first 100 crimes in the df_incidents dataframe
limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
import folium
```

```python
world_map = folium.Map(location=[56.130, -106.35], zoom_start=4, tiles='cartodbpositron')
world_map
```

```python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42
# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12, tiles='Cartodb dark_matter')
incidents = folium.map.FeatureGroup()
# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
# add incidents to map
sanfran_map.add_child(incidents)
```
